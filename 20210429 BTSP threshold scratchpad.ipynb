```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 10
active_input_index = int(dim/2)
input_activity = np.zeros(dim)
input_activity[active_input_index] = 1.
output_activity = np.copy(input_activity)
```

```python
hidden_FF_weights = np.random.uniform(0., 1., (dim, dim))
hidden_FB_weights = np.random.uniform(0., 1., (dim, dim))
```

```python
fig, axes = plt.subplots(1, 2)
cbar = axes[0].imshow(hidden_FF_weights)
plt.colorbar(cbar, ax=axes[0])
axes[1].imshow(hidden_FB_weights)
plt.colorbar(cbar, ax=axes[1])
fig.show()
```

```python
hidden_FF_activity = hidden_FF_weights.dot(input_activity)
hidden_FB_activity = hidden_FB_weights.dot(output_activity)
```

```python
hidden_FF_activity
```

```python
plt.figure()
th = 0.5
plt.plot(hidden_FF_activity, c='k')
plt.plot(hidden_FB_activity, c='r')
plt.plot(np.logical_and(hidden_FF_activity > th, hidden_FB_activity > th), c='b')
```
